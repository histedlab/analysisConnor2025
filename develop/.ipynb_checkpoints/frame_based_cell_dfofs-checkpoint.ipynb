### Get trial based responses

```python
def get_trial_resp_from_pix(cell_masks,trial_frames,iTrialVis,stimFrs,nTrials):
    'returns an array with a df/f value for each cell for each trial for each vis stim'
    dfof_cellTrials = np.zeros((nCell,nOris,nTrials))
    visTrials_stimFrs = trial_frames[iTrialVis, :, stimFrs[0]:stimFrs[1], :, :].mean(axis=2)#trial types, trials, frs, nRows, nRows (mean over frs)
    trials_BL = trial_frames[:, :, :stimFrs[0], :, :].mean(axis=(0,1,2))

    for iC in range(nCell):
        stim_f = visTrials_stimFrs[:, :, cell_masks[:,:,iC] == 1].mean(axis=2) #shape=oris,trials,pixels that = cell, avg across pix
        bl = trials_BL[cell_masks[:,:,iC] == 1].mean()
        dfof_cellTrials[iC, :, :] = (stim_f-bl)/bl*100
    
    return dfof_cellTrials
```

```python
dfof_cellTrials_pre = get_trial_resp_from_pix(cell_masks,trial_frames_pre,iTrialVis,stimFrs,nTrials)
dfof_cellTrials_post = get_trial_resp_from_pix(cell_masks,trial_frames_post,iTrialVis,stimFrs,nTrials)
```

```python
#get sem by sorting trials small to large
allTrialsV_pre_sorted = np.full((6,30),np.nan)
allTrialsV_post_sorted = np.full((6,30),np.nan)
for ori in range(nOris):
    allTrialsV_pre_sorted[ori,:] = np.sort(dfof_cellTrials_pre[16,ori,:])
    allTrialsV_post_sorted[ori,:] = np.sort(dfof_cellTrials_post[16,ori,:])
allTrialsV_change = allTrialsV_post_sorted-allTrialsV_pre_sorted
sem_trials = stats.sem(allTrialsV_change,axis=1)

sem_trials = np.concatenate((sem_trials,sem_trials[:1]))
avg_trials = np.concatenate((allTrialsV_change.mean(axis=1),allTrialsV_change.mean(axis=1)[:1]))
```

## Get avg trial from pix

```python
def get_avg_trial_from_pix(cell_masks,trial_frames,iTrialVis,nFrs):
    'returns an array with a df/f value for each cell for each trial for each vis stim'
    dfof_avgTrial_cell = np.zeros((nCell,nOris,nFrs))
    avgTrials_F = trial_frames[iTrialVis, :, :, :, :].mean(axis=1)#trial types, trials, frs, nRows, nRows (mean over frs)
    trials_BL = trial_frames[:, :, :stimFrs[0], :, :].mean(axis=(0,1,2))

    dfof_avgTrial_frs = (avgTrials_F-trials_BL)/trials_BL*100
    for iC in range(nCell):
        all_pix_for_cell = dfof_avgTrial_frs[:, :, cell_masks[:,:,iC] == 1]  #oris,frs,pix
        dfof_avgTrial_cell[iC, :, :] = all_pix_for_cell.mean(axis=2) #get the cell vals over oris and frames
    
    return dfof_avgTrial_cell
```

```python
dfof_avgTrial_cell_pre = get_avg_trial_from_pix(cell_masks,trial_frames_pre,iTrialVis,nFrs)
dfof_avgTrial_cell_post = get_avg_trial_from_pix(cell_masks,trial_frames_post,iTrialVis,nFrs)
```
